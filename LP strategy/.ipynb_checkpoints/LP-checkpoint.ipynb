## Objective function

Let $X$ be the numéraire, $Y$ be the risky asset.
Suppose liquidity provider(LP) provides liquidity on the range $[p_{i_l},p_{i_u}]$. On each $[_j,p_{j+1}]_{j=i_l}^{i_u - 1}$, the liquidity
at time $t$ is denoted by $L_j(t)$, i.e LP liquidity profile varing in time is a process
$L: \mathbb{Z} \times \mathbb{N} \times \Omega \to \mathbb{R}^+$.
We denote LP's position at time $t$ by $(X_t,Y_t)$, LP's wealth(mark to market price $S_t$) by $W_t = X_t+Y_t S_t$,
the fee in $X$ earned at time $t$ by $f_X(S_t,S_{t+1} ; L(t))$ , the fee in $Y$ earned at time $t$ by $f_Y(S_t,S_{t+1} ; L(t))$. 

At time $T$, the LP will liquidate his position. So he wants to maximize the value of all his asset(fee and wealth) at time $T$, which can be formulated as
the following:
$$
\begin{align}
&\sup_{\{L_j( \cdot) \}_{j=i_l}^{i_u - 1}} \mathbb{E}\big [ (X_T + Y_T S_T)e^{-rT} +
\sum^{T-1}_{t=0} e^{-r(t+1)}f_X(Z_t,Z_{t+1} ; L(t)) + e^{-r(t+1)}f_Y(Z_t,Z_{t+1}); L(t)) \cdot S_{t+1}  \big] \\
& =
\sup_{\{L_j( \cdot)\}_{j=i_l}^{i_u - 1} } \mathbb{E}\big[
W_T+\sum^{T-1}_{t=0}  e^{-r(t+1)}f_X(Z_t,Z_{t+1} ; L) +  e^{-r(t+1)}f_Y(Z_t,Z_{t+1}); L(t)) \cdot S_{t+1}  \big] \\
& =\sup_{\{L_j( \cdot)\}_{j=i_l}^{i_u - 1} } \mathbb{E}\big[  W_T+
\sum_{\{k: \tau^T_k \leq T\}} e^{-r(t+1)} f_X(Z_{\tau^T_k-1},Z_{\tau^T_k} ; L(t)) +  e^{-r(t+1)}f_Y(Z_{\tau^T_k-1},Z_{\tau^T_k}); L(t)) \cdot S_{\tau^T_k}  \big] 
\end{align}
$$
where $r \geq 0$ is the risk free rate and $\tau^T_k$ is the k-th time LP adjusts his liquidity determined by $L$. Note that for stable pair, we may
ignore the wealth part since the price should be back to the initial price.

If we have:

* The gas fee of changing liquidity isn't scaled with the size

* Liquidity deth doesn't affect pool price dynamics(e.g the arrival rate of traders)

* Not stable pair, price of risky asset is more like random walk rather than short period mean-reverting process.

* our optimization doesn't invole risk management(e.g above objective function consider only expected reward)

Then according to optimal stopping, the time withdraw liquidity is when the expected reward afterward is lower than current position value(wealth), in other words, the  earning ability of the current tick is sufficiently lower s.t it can't even earn the current position value. 

From the above, we can assume when LP decreases liquidity on a tick interval, he withdraw all liquidity.

The below plot shows tghe sublinear growth of cumulative fee on connstant LP range, i.e
$$
\lim_{n \to \infty} \frac{fee_t}{t} = 0
$$

## Pool mechanism


``` tick_price ``` : left endpoint of a tick as the representation of that tick. 

````python
gamma = 1-0.0001

token_0= 'USDC' # X numéraire

token_1 = 'USDT' # Y
class LP_position:
    def __init__(self,L={}):
        self.L_profile = L
        self.fee = np.zeros(2)
    def add_liquidity(self, tick_price, liquidity):
        '''
        liquidity: if >0,  add liquidity,  if <0, withdraw liquidity 
        '''
        if tick_price in self.L_profile.keys():
            if self.L_profile[tick_price]<= - liquidity:
                self.L_profile[tick_price] = 0
                print('not enough liquidity to withdraw')
            else:
                self.L_profile[tick_price]+= liquidity
            
        else:
            self.L_profile[tick_price] = liquidity

                
    def get_liquidity(self,tick_price):
        if ~(tick_price in self.L_profile.keys()):
            return 0
        return self.L_profile[tick_price]
    
    
    def plot_liquidity(self):
        plt.figure(figsize=(8,4))
        plt.bar(self.L_profile.keys() ,self.L_profile.values() , 
                width =1,align='edge',ec='black',color='lightblue')
        plt.title("liquidity profile")
        plt.xlabel("pool price")
        plt.ylabel("liquidity")

    
    def get_token_amounts(self, pool_price):
        sqrt_p = np.sqrt(pool_price)
        total_X = 0
        total_Y = 0
        ticks = np.sort(list(self.L_profile.keys()))
        for s in ticks:
            pl = tick_2_price(s,token_0,token_1)
            pr = tick_2_price(s+1,token_0,token_1)
            sqrt_pl = np.sqrt(pl)
            sqrt_pr = np.sqrt(pr)
            sqrt_p = np.sqrt(pool_price)
            L = self.L_profile[s]
            if pool_price <= pl:
                total_Y += L * (sqrt_pr - sqrt_pl) / (sqrt_pr * sqrt_pl)
            elif pool_price >= pr:
                total_X += L * ( 1/sqrt_pl - 1/sqrt_pr)
            else:
                total_Y += L * (sqrt_pr - sqrt_p) / (sqrt_p * sqrt_pr)
                total_X += L * (sqrt_p - sqrt_pl)
        return np.array([total_X, total_Y])

    def get_position_value(self, market_price):
        X, Y = self.get_token_amounts(market_price)
        return X + Y * market_price  
    
    
    def add_fee(self,price_n,price_o):
        dX , dY  = self.get_token_amounts(price_n) - self.get_token_amounts(price_o)
        fee_X , fee_Y = (0,0)
        if price_n <= price_o:
            fee_Y = (1-gamma)/gamma *dY
        
        else:
            fee_X = (1-gamma)/gamma *dX
        
        self.fee+= np.array([fee_X,fee_Y])
````

````python
decimal_dict ={  'WETH':18,'DAI':18, 'USDT':6 , 'USDC':6 }
def tick_2_price(tick,token_0=token_0,token_1=token_1):
    decimal_0 = decimal_dict[token_0]
    decimal_1 = decimal_dict[token_1]


    return 1.0001**tick * 10 ** (decimal_1 - decimal_0)

def price_2_tick(price,token_0=token_0,token_1=token_1):
    decimal_0 = decimal_dict[token_0]
    decimal_1 = decimal_dict[token_1]


    if price == 0:
        return -np.inf

    tick = math.log(price, 1.0001) - (decimal_1 - decimal_0) * math.log(10, 1.0001)
    return round(tick)
````

``` which_tick```: Return the activated tick w.r.t the given pool price. 

```fee_per_L```: Return a dictionary where the keys are ticks that are onced activated w.r.t a price array and values are tuples corresponding to the fees paid in token 0 and token 1.

```fee_per_L_in_X```: Similar to ```fee_per_L```, but the dictionary values
are evaluated in token 0(X asset) using the tick price or mid price. 


```cost_per_L```: Compute the amount of tokens needed to deposit for initiating
a unit liquidity position under current pool price. The token 1 value is evaluated in ```ref_price```.

````python
def which_tick(price,token_0=token_0,token_1=token_1,show_price=False):
    decimal_0 = decimal_dict[token_0]
    decimal_1 = decimal_dict[token_1]
    tick = int(math.log(price/10**(decimal_1-decimal_0),1.0001    ))
    if show_price:
        return tick_2_price(tick)
    return tick

# fee in X and Y per liquidity based on history price
def fee_per_L(price_array,LP_range=None):
    max_tick = which_tick(price_array.max())
    min_tick  =which_tick(price_array.min())
    ticks = np.arange(min_tick,max_tick+1)
    fee_on_tick= { tk: np.zeros(2) for tk in ticks }
    for t,p_new in enumerate(price_array):
        if t==0:
            continue 
        p_old = price_array[t-1]
        tick_old = which_tick(p_old)
        tick_new = which_tick(p_new)
        if tick_old == tick_new:
            # buy Y pay X
            if p_old < p_new:
                fee_on_tick[tick_old][0] +=  (1-gamma)/gamma*(np.sqrt(p_new) - np.sqrt(p_old)) 
            # Sell Y pay Y
            else:
                fee_on_tick[tick_old][1] += (1-gamma)/gamma*(1/np.sqrt(p_new) - 1/np.sqrt(p_old)) 
        else:
            if p_old <p_new:
                for tk in range(tick_old , tick_new+1):
                    if tk == tick_old:
                        p_i = p_old
                        p_f = tick_2_price(tk+1)
                    elif tk==tick_new:
                        p_i = tick_2_price(tk)
                        p_f = p_new
                    else:
                        p_i = tick_2_price(tk)
                        p_f = tick_2_price(tk+1)
                        
                    fee_on_tick[tk][0] +=  (1-gamma)/gamma*(np.sqrt(p_f) - np.sqrt(p_i)) 
            else:
                for d in range(tick_old-tick_new+1):
                    tk = tick_old - d
                    if tk == tick_old:
                        p_i = p_old
                        p_f = tick_2_price(tk)
                    elif tk==tick_new:
                        p_i = tick_2_price(tk)
                        p_f = p_new
                    else:
                        p_i = tick_2_price(tk)
                        p_f = tick_2_price(tk-1)
                        
                    fee_on_tick[tk][1] += (1-gamma)/gamma*(1/np.sqrt(p_f) - 1/np.sqrt(p_i)) 

    
    if LP_range  is not None:
        # intersection part
        original_ticks = np.intersect1d( ticks,  LP_range) 
        # in LP range but not price range
        new_ticks = np.setdiff1d( LP_range,ticks  )
        fee_on_tick={ tk:fee_on_tick[tk] for tk in original_ticks  }
        for tk in new_ticks:
            fee_on_tick[tk]= np.zeros(2)
        
    return fee_on_tick
def fee_per_L_in_X(fee_on_tick,how=None):
    fee_on_tick_in_X = {tk:0 for tk in fee_on_tick.keys()}
    for tk, fees in fee_on_tick.items():
        # geometic mean
        if how =='mid': 
            price =  (tick_2_price(tk+1)* tick_2_price(tk))**(1/2)
        elif how=='left':
            price =   tick_2_price(tk)
        elif how=='right':
            price = tick_2_price(tk+1)
        elif how==None:
            price = 1
        fee_on_tick_in_X[tk] = fees[0]+price*fees[1]      
    return fee_on_tick_in_X
````

````python
# using p_new as ref price, not very  different from fee_per_L_in_X 
def fee_per_L_in_X2(price_array):
    max_tick = which_tick(price_array.max())
    min_tick  =which_tick(price_array.min())
    ticks = np.arange(min_tick,max_tick+1)
    fee_on_tick_in_X = {tk:0 for tk in ticks}
    for t,p_new in enumerate(price_array):
        if t==0:
            continue 
        p_old = price_array[t-1]
        tick_old = which_tick(p_old)
        tick_new = which_tick(p_new)
        if tick_old == tick_new:
            # buy Y pay X
            if p_old < p_new:
                fee_on_tick_in_X[tick_old] +=  (1-gamma)/gamma*(np.sqrt(p_new) - np.sqrt(p_old)) 
            # Sell Y pay Y
            else:
                fee_on_tick_in_X[tick_old] += ((1-gamma)/gamma*(1/np.sqrt(p_new) - 1/np.sqrt(p_old)))*p_new 
        else:
            if p_old <p_new:
                for tk in range(tick_old , tick_new+1):
                    if tk == tick_old:
                        p_i = p_old
                        p_f = tick_2_price(tk+1)
                    elif tk==tick_new:
                        p_i = tick_2_price(tk)
                        p_f = p_new
                    else:
                        p_i = tick_2_price(tk)
                        p_f = tick_2_price(tk+1)
                        
                    fee_on_tick_in_X[tk] +=  (1-gamma)/gamma*(np.sqrt(p_f) - np.sqrt(p_i)) 
            else:
                for d in range(tick_old-tick_new+1):
                    tk = tick_old - d
                    if tk == tick_old:
                        p_i = p_old
                        p_f = tick_2_price(tk)
                    elif tk==tick_new:
                        p_i = tick_2_price(tk)
                        p_f = p_new
                    else:
                        p_i = tick_2_price(tk)
                        p_f = tick_2_price(tk-1)
                        
                    fee_on_tick_in_X[tk] += ((1-gamma)/gamma*(1/np.sqrt(p_f) - 1/np.sqrt(p_i)))*p_new

    
    return fee_on_tick_in_X
````

````python
# ini cost per liquidity
def cost_per_L(tick,pool_price, ref_price=1):
    '''
    tick: the tick to provide 1 liquidity 
    '''
    if ref_price is None:
        ref_price = pool_price
    tick_now = which_tick(pool_price)
    p = tick_2_price(tick)
    if tick < tick_now:
        cost = np.sqrt( tick_2_price(tick+1) ) - np.sqrt( p)
    elif tick > tick_now:
        cost = ref_price*( 1/np.sqrt(p) - 1/np.sqrt(tick_2_price(tick+1) ) )  
    else:
        X= np.sqrt(pool_price ) - np.sqrt(p)
        Y= 1/ np.sqrt( pool_price) - 1/ np.sqrt( tick_2_price(tick+1) )
        cost =  X+ ref_price*Y
    return cost


````

## Model-based strategy

### Markov


````python
def MC(df,separation_time):
    price_array = df[df['datetime']<= separation_time ]['price'].values
    tick_states =np.sort(np.unique([ which_tick(price) for price in price_array ]))
    tick_dict= { value:index for index,value in enumerate(tick_states) }
    tran_mat  =np.zeros( (len(tick_states),len(tick_states)  ) )
    for t,price in enumerate(price_array[1:]):     
        tick_prev = which_tick(price_array[t-1])
        tick_current = which_tick(price)
        index_prev = tick_dict[tick_prev]
        index_current = tick_dict[tick_current]
        
        tran_mat[index_prev,index_current] += 1
    tran_count = tran_mat.sum(axis=1)
    
    return np.divide(tran_mat,tran_count[:,np.newaxis]),tick_states
````

````python
# coarsed granning mc, for visualization
def MC_CG(df,separation_time,n_groups=20):
    price_array = df[df['datetime']<= separation_time ]['price'].values
    ticks =np.sort(np.unique([ which_tick(price) for price in price_array ]))
    n = len(ticks) 
    tk = ticks.reshape(-1, 1) 
    kmeans = KMeans(n_clusters=n_groups,algorithm='elkan').fit(tk)
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_ 
    order = np.argsort(centroids.flatten())
    label_map = {old_label: new_label for new_label, old_label in enumerate(order)}
    labels = np.array([label_map[label] for label in labels])

    # ticks to labels mapping
    tick_dict= { tk:lab for tk,lab in zip(ticks,labels) }
    
    tran_mat  =np.zeros( (n_groups,n_groups ) )


    for t,price in enumerate(price_array[1:]):     
        tick_prev = which_tick(price_array[t-1])
        tick_current = which_tick(price)
        lab_prev = tick_dict[tick_prev]
        lab_current = tick_dict[tick_current]
        
        tran_mat[lab_prev,lab_current] += 1
    tran_count = tran_mat.sum(axis=1)
    
    return np.divide(tran_mat,tran_count[:,np.newaxis]),tick_dict
````

## Model-free/Data-driven strategy

### Simple static
Given initial pool price $p_0$ and initial capital $W$, the LP profile has the constraint

$$
W = X + p_0 Y
=\frac{1}{2} \int_{0}^{p_0} L(z) z^{-\frac{1}{2}} dz +
\frac{1}{2} \int_{p_0}^\infty p_0 L(z) z^{-\frac{3}{2}} dz 
$$

and we want to maximize the expected fee
$$
\max_L  \mathbb{E} \big[ cumfee(L,T,r) \big]
$$
where $T$ is the terminal time and $r$ is risk free rate.

1. Return weighted ETF: assign weight according to return of that tick

````python
# if traning period too short, price change little
def static_LP(df,separate_time = datetime(2023,1,1),ini_cap=100,width=150):
    
    # excluding outliers, focus on the normal price range 0.9~1.1
    df = df[df['price'].between(0.9,1.1) ].copy()
    
    price_array_IS = df[df['datetime']<=separate_time]['price'].values
    price_array_OS = df[df['datetime']>separate_time]['price'].values

    #fee per cost on each tick that activated
    fee_dict = fee_per_L_in_X(fee_per_L(price_array_IS))
    p0 = price_array_IS[0]
    # divided by cost
    fee_dict = { tk:value/cost_per_L(tk,p0)    for tk,value in fee_dict.items()  }
    
    #most profitable ticks
    tick_center = max(fee_dict, key=fee_dict.get)
    
    LP_range= np.arange(tick_center-width,tick_center+width)

    # choose intersection of lp range and ticks from data
    LP_range = np.intersect1d(LP_range,np.array( list(fee_dict.keys()) )  )
    
    S = np.sum([   fee_dict[tk] for tk in LP_range]) 
    LP_profile = { tk:ini_cap*fee_dict[tk]/S * 1/cost_per_L(tk,  price_array_OS[0])
                  for tk in LP_range   }
    F = fee_per_L(price_array_OS,LP_range)
    F = { tk:F[tk]*LP_profile[tk] for tk in LP_range  }
    total_X = np.sum([ fee[0]   for fee in  F.values() ])
    total_Y = np.sum([ fee[1]   for fee in  F.values() ])
    return (total_X,total_Y),LP_profile
````

2. All-in strategy: Put all capital in the tick that has the most return according to backtest

````python
def static_LP2(df,separate_time = datetime(2023,1,1),ini_cap=100):
    df = df[df['price'].between(0.9,1.1) ].copy()

    price_array_IS = df[df['datetime']<=separate_time]['price'].values
    price_array_OS = df[df['datetime']>separate_time]['price'].values

    fee_dict = fee_per_L_in_X(fee_per_L(price_array_IS))
    p0 = price_array_IS[0]
    fee_dict = { tk:value/cost_per_L(tk,p0)    for tk,value in fee_dict.items()  }
    
    #most profitable ticks
    tick_center = max(fee_dict, key=fee_dict.get)
    p_l =tick_2_price(tick_center)
    p_u = tick_2_price(tick_center+1)
    L = 1/cost_per_L(tick_center, price_array_OS[0]  )*ini_cap
    fee = np.zeros(2) 
 
    for t in range(1,len(price_array_OS)):
        p_new = price_array_OS[t]
        p_old = price_array_OS[t-1]

        if p_old<p_new:
            f =  (1-gamma )/gamma * L* (np.sqrt( min(p_u, p_new)) - np.sqrt(max(p_l,p_old))   )
            fee[0] += max(0,f)
            
        else:
            f = (1-gamma )/gamma *L* (1/np.sqrt(max(p_l,p_new)) - 1/np.sqrt( min(p_u, p_old)) )
            fee[1] += max(0,f)
        
    return fee,{tick_center:L}
````

### Parametric optimization

```static_LP_normal```: 

* ```max_tick```: If true, use the in sample most profitable tick as center of     normal LP profile.If not, use tick 0 as center(in this case, ). 

* ```std```: Standard deviation of liquidity profile.

* ```width```: Width of truncated normal, corresponding LP range is 
               $(P_{center}-width*std, P_{center}+width*std) $ where $P_{center}$                 is the IS most profitable tick or tick 0 depending on                              ```max_tick``` setting.

````python
def static_LP_normal(df,separate_time = datetime(2023,1,1),max_tick=True,ini_cap=100,std=5,width=1):
    df = df[df['price'].between(0.9,1.1) ].copy()

    price_array_IS = df[df['datetime']<=separate_time]['price'].values
    price_array_OS = df[df['datetime']>separate_time]['price'].values
    if max_tick == True:
        fee_dict = fee_per_L_in_X(fee_per_L(price_array_IS))
        p0 = price_array_IS[0]
        fee_dict = { tk:value/cost_per_L(tk,p0)  for tk,value in fee_dict.items()  }
        tick_center = max(fee_dict, key=fee_dict.get)
    # use price=1 as center( tick = 0)
    else:
        tick_center = 0
    LP_range =np.arange(int(tick_center-width*std),int(tick_center+width*std)+1 )

    
    #truncated normal 
    weight = {x:scipy.stats.norm.pdf(x,loc=tick_center,scale=std)  for x in LP_range}
    S = np.sum(list( weight.values() ))
    #most profitable ticks
    LP_profile = { tk:ini_cap*weight[tk]/S * 1/cost_per_L(tk,price_array_OS[0])
                  for tk in LP_range   }
    
    F = fee_per_L(price_array_OS,LP_range)
    F = { tk:F[tk]*LP_profile[tk] for tk in LP_range  }
    total_X = np.sum([ fee[0]  for fee in  F.values() ])
    total_Y = np.sum([ fee[1]  for fee in  F.values() ])
    return (total_X,total_Y),LP_profile
````

````python
# remove outliers
def fee_animation(df):
    dff=  df[df['price'].between(0.9,1.1) ].copy()
    price_array = dff['price'].values
    max_tick = which_tick(price_array.max())
    min_tick  =which_tick(price_array.min())
    ticks = np.arange(min_tick,max_tick+1)
    fee_on_tick= { tk: [] for tk in ticks }
    for t,p_new in enumerate(price_array):
        if t==0:
            continue 
        p_old = price_array[t-1]
        tick_old = which_tick(p_old)
        tick_new = which_tick(p_new)
        if tick_old == tick_new:
            # buy Y pay X
            if p_old < p_new:
                fee_on_tick[tick_old][0] +=  (1-gamma)/gamma*(np.sqrt(p_new) - np.sqrt(p_old)) 
            # Sell Y pay Y
            else:
                fee_on_tick[tick_old][1] += (1-gamma)/gamma*(1/np.sqrt(p_new) - 1/np.sqrt(p_old)) 
        else:
            if p_old <p_new:
                for tk in range(tick_old , tick_new+1):
                    if tk == tick_old:
                        p_i = p_old
                        p_f = tick_2_price(tk+1)
                    elif tk==tick_new:
                        p_i = tick_2_price(tk)
                        p_f = p_new
                    else:
                        p_i = tick_2_price(tk)
                        p_f = tick_2_price(tk+1)
                        
                    fee_on_tick[tk][0] +=  (1-gamma)/gamma*(np.sqrt(p_f) - np.sqrt(p_i)) 
            else:
                for d in range(tick_old-tick_new+1):
                    tk = tick_old - d
                    if tk == tick_old:
                        p_i = p_old
                        p_f = tick_2_price(tk)
                    elif tk==tick_new:
                        p_i = tick_2_price(tk)
                        p_f = p_new
                    else:
                        p_i = tick_2_price(tk)
                        p_f = tick_2_price(tk-1)
                        
                    fee_on_tick[tk][1] += (1-gamma)/gamma*(1/np.sqrt(p_f) - 1/np.sqrt(p_i)) 
````

````python
def fee_per_L(price_array,LP_range=None):

    
    if LP_range  is not None:
        # intersection part
        original_ticks = np.intersect1d( ticks,  LP_range) 
        # in LP range but not price range
        new_ticks = np.setdiff1d( LP_range,ticks  )
        fee_on_tick={ tk:fee_on_tick[tk] for tk in original_ticks  }
        for tk in new_ticks:
            fee_on_tick[tk]= np.zeros(2)
        
    return fee_on_tick
def fee_per_L_in_X(fee_on_tick,how=None):
    fee_on_tick_in_X = {tk:0 for tk in fee_on_tick.keys()}
    for tk, fees in fee_on_tick.items():
        # geometic mean
        if how =='mid': 
            price =  (tick_2_price(tk+1)* tick_2_price(tk))**(1/2)
        elif how=='left':
            price =   tick_2_price(tk)
        elif how=='right':
            price = tick_2_price(tk+1)
        elif how==None:
            price = 1
        fee_on_tick_in_X[tk] = fees[0]+price*fees[1]      
    return fee_on_tick_in_X
````
