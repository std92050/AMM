```python
import matplotlib.pyplot as plt
import numpy as np
```

This note is about how to adjust coordinate axis ticks and ticklabels .

```python
fig= plt.figure(figsize=(7,7))
#3d graph
ax = plt.subplot( projection = "3d")

x = np.arange(10)
y = 3*np.arange(10)+0.03

ax.bar3d(x,y,0,dx=0.5,dy=0.5 ,dz=0.1)


# create the ticks and labels we want to add
new_ticks_dict = { 5.3013 :'8=D' , 3.6 :'dick' , 8 : "dd"}

# update x axis ticks by adding ticks part of dict 
new_ticks = np.array(  list(  new_ticks_dict.keys()   ) )
x_ticks = np.append(ax.get_xticks() , new_ticks )
ax.set_xticks(x_ticks)

# get current ticklables (now already includes new ticks)
ticklabels = ax.get_xticklabels()
# run over all labels
for index,lab  in enumerate(ticklabels):
    
    # tick of that label 
    lab_tick = x_ticks[index]
    lab.set_rotation(45)  # rotate
    # change the label of new tick we added
    if lab_tick in new_ticks :
        lab.set_text(new_ticks_dict[lab_tick]) #new label
        lab.set_fontsize(8)  # fontsize      
        lab.set_color('r')    #color  
ax.set_xticklabels(ticklabels)
ax.plot( 1*np.ones(10) , y , color = 'r' )

#ax.view_init(azim = 40 ) #rotation the 3d graph (default azim=-45)
```
