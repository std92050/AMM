```python
import string
book=open('b2')
hist={}
t=[]
def f(g):
    for word in g.split():
        word=word.strip(string.punctuation)
        word=word.lower()
        hist[word]=hist.get(word,0)+1
    
for line in book:
    f(line)
for key,value in hist.items():
    t.append((value,key))
    t.sort(reverse=True)
    
    
```
