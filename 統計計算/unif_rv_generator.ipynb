# LCG Linear congruential generators
This method create a uniform pesudo random number in \\([0,1]\\).
Starting from a given seed \\(\in [0,m-1] \\) , the next put value will
be \\(a*seed+c\\).Iteratively,the sequence of numbers divided by m generated by this method are approximately uniform r.vs on \\([0,1]\\) 

```python

def LCG(m,a,c,seed):
    X=np.array([seed])
    for i in range(10*m):
        x=a*X[i]+c
        X=np.append(X,x%m)
    X=np.delete(X,0)
    return X/m

LCG(100,18,0,3)
```

```python
#another def of lcg function ★generator的寫法 
def lcg(modulus, a, c, seed):
    """Linear congruential generator."""
    while True:
        seed = (a * seed + c) % modulus
        yield seed/modulus
x=lcg(2**31-1,7**5,0,3)
for i in x:
    print(i)
```

# MRG Multiple-recursive generator
This is another method to sample from a uniform r.v.
Different from lcg,the output value depends on more than one previous 
output value. It is said to be of $order$ $k$ if the output value depends
on previous k values.

```python
def MRG(m,a_vector,seed_vector):
    k=len(a_vector)
    S=seed_vector
    for i in range(10*m):
        x=np.dot(a_vector,S[i:k+i])%m
        S=np.append(S,x)
    return S/m
MRG(10,np.array([1,2]),np.array([2,2]))
```

# MCG Matrix congruential generator
This method yields a vector sampled from a unifom r.v vector whose components are in $[0,1]$.
The output vector is $v_{n+1}$=$Av_n$ , where $A$ is a given matrix and $v_n$ is previous output vector.

```python
def MCG(m,A,seed_vector):
    v_0=A@seed_vector%m
    Sample=np.array([v_0])
    
    for i in range(10*m):
        
        v=A@Sample[i]%m
        Sample=np.vstack((Sample,v))
    return Sample
MCG(100,np.array([[2,3,4],[1,1,1],[0,0,0]]),np.array([23,3,99]))
```
