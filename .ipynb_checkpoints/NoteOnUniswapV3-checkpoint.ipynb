## Notations

- $x_t$: number of risky asset, say ETH, at time $t$ in the pool 
- $y_t$: number of numeraire, say USDC, at time $t$ in the pool
- $L_t$: liquidity depth as a function of infinitesimal rate of exchange $P$ at time $t$
- $P_t = -\frac{\p y_t}{\p x_t}$: infinitesimal rate of exchange at time $t$

## Liquidity as a function of infinitesimal rate of exchange - the *liquidity profile*

Assume that $L = L(P)$, i.e., liquidity depends on the infinitesimal rate of exchange $P$, which we shall refer to as the <font color=blue>*liquidity profile*</font>. 

## Converting liquidity profile to reserve curve

The pool reserve $(x,y)$ is given in terms of the infinitesimal rate of exchange $P$ as

\begin{eqnarray*}
&& \p x = -\frac{L(P)}2 P^{-\frac32} dP \quad \Longrightarrow \quad x(P) = \frac12 \int_P^\infty L(p) p^{-\frac32} dp \\
&& \p y = \frac{L(P)}2 P^{-\frac12} dP \quad \Longrightarrow \quad y(P) = \frac12 \int_0^P L(p) p^{-\frac12} dp \\
\end{eqnarray*}

Apparently, we have 

$$
\frac{\p y}{\p x} = \frac{\frac{\p y}{\p P}}{\frac{\p x}{\p P}} = -P,
$$

In other words, the infinitesimal rate of exchange $P$ is equal to the negative slope of tangent of the reserve curve. 

## Alternative formulation

By applying integration by parts, we can rewrite $x$ and $y$, as a function of $P$, as

$$
x(P) = \frac12 \int_P^\infty L(p) p^{-\frac32} dp = - \int_P^\infty L(p) d\left(\frac1{\sqrt p}\right) 
= \frac{L(P)}{\sqrt P} + \int_P^\infty \frac1{\sqrt p}dL(p)
$$

and 

$$
y(P) = \frac12 \int_0^P L(p) p^{-\frac12} dp = \int_0^P L(p) d\sqrt p
= \sqrt P L(P) - \int_0^P \sqrt p dL(p)
$$

#### Note
- Mathematically, we can regard $dL(p)$ as a ($\sigma$-finite) positive measure over $[0, \infty)$. Thus, $L$ is the associated cdf for the measure $dL$.

## Illustration for the conversion of liquidity profile to reserve 

For example, if the liquidity profile is given by 

$$
L(p) = k_1 \1_{[a, b)}(p) + k_2 \1_{[b, c)}(p)
$$ 

for some positive constants $k_1, k_2$. Then, we have 

$$
dL(p) = k_1 \delta(p - a) dp + (k_2 - k_1)\delta(p - b) dp - k_2 \delta(p-c) dp.
$$

By simply carrying out the integrations, one obtains 

- for $P \geq c$, apparently
$$
x = 0;
$$
- for $b \leq P < c$
$$
x = \frac{L(P)}{\sqrt P} + \int_P^\infty \frac1{\sqrt p}dL(p) = \frac{k_2}{\sqrt P} - \frac{k_2}{\sqrt c} = k_2 \left(\frac1{\sqrt P} - \frac1{\sqrt c}\right)
$$
- for $a \leq P < b$
\begin{eqnarray*}
x &=& \frac{L(P)}{\sqrt P} + \int_P^\infty \frac1{\sqrt p}dL(p) = \frac{k_1}{\sqrt P} + (k_2 - k_1) \left(\frac1{\sqrt b} - \frac1{\sqrt c}\right) - \frac{k_2}{\sqrt c} \\
&=& k_1 \left(\frac1{\sqrt P} - \frac1{\sqrt b}\right) + k_2 \left(\frac1{\sqrt b} - \frac1{\sqrt c}\right)
\end{eqnarray*}
- for $P < a$
\begin{eqnarray*}
x &=& \frac{k_1}{\sqrt a} + \frac{k_2 - k_1}{\sqrt b} - \frac{k_2}{\sqrt c}
\end{eqnarray*}

Similarly, as for $y$ we have 
- for $P \leq a$, apparently
$$
y = 0;
$$
- for $a < P \leq b$
$$
y = \sqrt P L(P) - \int_0^P \sqrt p dL(p) = k_1 \left(\sqrt P - \sqrt a \right)
$$
- for $b < P \leq c$
\begin{eqnarray*}
y &=& \sqrt P L(P) - \int_0^P \sqrt p dL(p) = k_2 \sqrt P - k_1 \sqrt a - (k_2 - k_1)\sqrt b \\
&=& k_1 \left(\sqrt b - \sqrt a\right) + k_2 \left(\sqrt P - \sqrt b\right)
\end{eqnarray*}
- for $P > c$
\begin{eqnarray*}
y &=& -k_1 \sqrt a - (k_2 - k_1)\sqrt b + k_2\sqrt c = k_1\left(\sqrt b - \sqrt a\right) + k_2\left(\sqrt c - \sqrt b\right)
\end{eqnarray*}

### Create a `python` `class` for ploting liquidty profile and reserve curve

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as ss

class LiquidityReservePlot():
    '''
    L = liquidity profile
    range/interval for P = [start=0, end=10] 
    '''
    def __init__(self, L, start=0.00001, end=10):
        self.L, self.start, self.end = L, start, end
        self.x1 = lambda p: self.L(p)/2/p**1.5
        self.y1 = lambda p: self.L(p)/2/np.sqrt(p)

        dP = 0.01
        P = np.arange(self.start, self.end, dP)
        
        # numerical intergration
        x = (self.x1(P).sum() - self.x1(P).cumsum())*dP
        y = self.y1(P).cumsum()*dP
        self.x, self.y = x, y
        
        #plot
        plt.figure(figsize=(12, 8))

        plt.subplot(2, 2, 1)
        plt.plot(P, self.L(P))
        plt.title('Liquidity profile', fontsize=18)
        plt.xlabel(r'$P$', fontsize=15)
        plt.ylabel(r'$L$', fontsize=15)

        plt.subplot(2, 2, 2)
        plt.plot(x, y)
        plt.title('Reserve curve', fontsize=18)
        plt.hlines(y=0, xmin=0, xmax=x.max(), color='grey', ls='dotted')
        plt.vlines(x=0, ymin=0, ymax=y.max(), color='grey', ls='dotted')
        plt.xlabel(r'$x$', fontsize=15)
        plt.ylabel(r'$y$', fontsize=15)

        plt.subplot(2, 2, 3)
        plt.plot(P, x)
        plt.xlabel(r'$P$', fontsize=15)
        plt.ylabel(r'$x$', fontsize=15)

        plt.subplot(2, 2, 4)
        plt.plot(P, y)
        plt.xlabel(r'$P$', fontsize=15)
        plt.ylabel(r'$y$', fontsize=15);
    
    def __call__(self):
        pass
```

### Constant  liquidity profile

```python
# constant liquidity profile
L = lambda p: 100*(p > 0)

ex1 = LiquidityReservePlot(L, end=10_000);
```

```python
start, end = 50, 10000
plt.plot(ex1.x[start:end], ex1.y[start:end]);
```

### Piecewise  constant liquidity  profile, finitely  supported

```python
a, b, c = 1, 2, 3
k1, k2 = 10, 50
L = lambda p: k1*(p >= a) + (k2 - k1)*(p >= b) - k2*(p > c)

LiquidityReservePlot(L);
```

### Continuous liquidity profile

```python
L = lambda p: 100*ss.chi2.pdf(p, df=3)
LiquidityReservePlot(L, end=15);
```

## Time dependent liquidity profile

If the liquidity profile $L$ depends on time, it follows that

\begin{eqnarray*}
&& x_t = \frac12 \int_P^\infty L_t(p)p^{-\frac32} dp \\
&& y_t = \frac12 \int_P^\infty L_t(p)p^{-\frac12} dp \\
&& \frac{\p y_t}{\p x_t} = -P.
\end{eqnarray*}

Thus,

\begin{eqnarray*}
&& dx_t(P) = \frac12 \int_P^\infty dL_t(p)p^{-\frac32} dp \\
&& dy_t(P) = \frac12 \int_P^\infty dL_t(p)p^{-\frac12} dp \\
&& d\left(\frac{\p y_t}{\p x_t}\right) = d\left(-P\right) = 0.
\end{eqnarray*}

The set of equations above convert the evolution of liquidity profile to the evolution of reserve curve. 

#### Note
- The infinitesimal rate of exchange $-\frac{\p y_t}{\p x_t}$ does not depend on time. 
- One may model the evolution of $L_t$ by an SPDE. For example, 
$$
dL_t = \cA_t L_t dt + \sigma L_t dW_t
$$
where $\cA_t$ is a second order parabolic/elliptic differential operator, like a stochastic heat equation.

## Trader trades at rate $u_t$

When trader trades, he must trade along the reserve curve.
Assume the liquidity profile $L$ remain unchanged, the trader trades at rate $u_t$, i.e., $dx_t = u_tdt$. 

Recall that 

$$
x = \frac12 \int_P^\infty L(p)p^{-\frac32} dp
$$

it follows that

$$
x_t = \frac12 \int_{P_t}^\infty L(p)p^{-\frac32} dp.
$$

Thus, if trader trades at the rate $u_t$, i.e., $dx_t = u_t dt$, then by differentiate the equation above with respect to $t$ on both sides we obtain

$$
u_t dt = dx_t = -\frac12 L(P_t)P_t^{-\frac32} dP_t
$$


Thus, if trader trades at the rate $u_t$, then the evolution of $P_t$ is given by the ODE

$$
-\frac12 L(P_t)P_t^{-\frac32} dP_t = u_t dt
$$

or equivalently, 

$$
dP_t = \frac{-2 P_t^{\frac32}}{L(P_t)} u_t dt 
$$


Likewise, for $y_t$ we have 

$$
y_t = \frac12 \int_0^{P_t} L(p)p^{-\frac12} dp.
$$

By differentiate the equation above with respect to $t$ on both sides we obtain

$$
dy_t = \frac12 L(P_t)P_t^{-\frac12} dP_t = -P_t \left\{-\frac12 L(P_t)P_t^{-\frac32} dP_t \right\} = -P_t u_t dt
$$


In summary, assume the liquidity profile stays steady, if the trader trades at rate $u_t$, i.e., $dx_t = u_t dt$, the evolutions of the pool and the infinitesimal rate of exchange are governed by 

\begin{eqnarray*}
&& dx_t = u_t dt \\
&& dy_t = -P_t u_t dt \\
&& dP_t = \frac{-2 P_t^{\frac32}}{L(P_t)} u_t dt 
\end{eqnarray*}

## Putting things together
## Market mechanism - without transaction cost

If the trader is trading at rate $u_t$ in time dependent liquidty profile $L_t(\cdot)$, the evolutions of the pool and the infinitesimal rate of exchange are given by 

\begin{eqnarray*}
dx_t &=& \frac12 \int_{P_t}^\infty dL_t(p)p^{-\frac32} dp + u_t dt \\
dy_t &=& \frac12 \int_0^{P_t} dL_t(p)p^{-\frac12} dp - P_t u_t dt \\
dP_t &=& \frac{-2 P_t^{\frac32}}{L_t(P_t)} u_t dt 
\end{eqnarray*}

Let $\xp = \frac{\p x}{\p P}$ and $\yp = \frac{\p y}{\p P}$. We may also consider the evolution for $\xp$ and $\yp$

\begin{eqnarray*}
d\xp_t &=& -\frac12 dL_t(P_t)P_t^{-\frac32} + \p_P  u_t dt \\
d\yp_t &=& \frac12 dL_t(P_t)P_t^{-\frac12} - (u_t + P_t \p_Pu_t) dt \\
dP_t &=& \frac{-2 P_t^{\frac32}}{L_t(P_t)} u_t dt 
\end{eqnarray*}
