```python
import sympy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

```python
def A(x,n,p):
    X=stats.Binomial('X',n,p)
    return stats.P(X>=x)-stats.P(X>x)
#這個function會給出一個B(n,p)的r.v是x的機率
```

```python
A(6,4,0.5)
```

```python
X=np.linspace(10,100,10)
Y=1/np.linspace(1,10,10)
#Z=np.array(A(10,int(i),j) for i in X for j in Y).reshape(10,10)
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z);
```
