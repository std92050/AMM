```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import norm , cauchy
```

```python
x=np.linspace(-5,5,9000)
f1=norm.pdf(x,0,0.5**2)
f2=norm.pdf(x,0,1)
f3=norm.pdf(x,0,4)
```

```python
plt.figure(figsize=(15,10))
fig,ax=plt.subplots(1,1)
ax.plot(x,f1,'r-',label='σ=0.5')
ax.plot(x,f2,'g-',label='σ=1')
ax.plot(x,f3,'b-',label='σ=2')
ax.legend(loc=1)
plt.savefig('gaussian.jpg')
plt.show()
```

```python
g1=cauchy.pdf(x,0,0.5**0.5)
g2=cauchy.pdf(x,0,1)
g3=cauchy.pdf(x,0,2**0.5)
```

```python
plt.figure(figsize=(15,10))
fig,ax=plt.subplots(1,1)
ax.plot(x,g1,'r-',label='γ=0.5')
ax.plot(x,g2,'g-',label='γ=1')
ax.plot(x,g3,'b-',label='γ=2')
ax.legend(loc=1)
plt.savefig('cauchy.jpg')
plt.show()
```
