Let $\{X_n\}_{n \geq 1}$ i.id with distribution
$$
P(X_1 =1) = P(X_1 = -1) = \frac{1}{2}
$$
Find the smallest $a \in \mathbb{R}$ s.t the series $\sum_{n=1}^\infty \frac{X_n}{n^a}$ converges
a.s .   ( clearly  $a > 0$ if it really exists )

In this note , we will run Monte-Carol to test the convergence of the series over a wide range of $a$ . Given $a > 0$ , let $Y_n = \frac{X_n}{n^a}$ .

```python
def Y(a,n):
    U = np.random.rand()
    if U > 0.5:
        return 1/n**a
    else :
        return -1/n**a
def S(a,N):
    S=np.zeros(N)
    sum = 0
    for n in range(N):
        sum += Y(a,n+1)
        S[n] = sum 
    return S
```

```python
plt.plot(S(1 , 500000) , label = "a = 1")
plt.plot(S(0.7 , 500000)  ,label = "a = 0.7")
plt.plot(S(0.6 , 500000)  ,label = "a = 0.6")
plt.plot(S(0.5 , 500000)  ,label = "a = 0.5")
plt.plot(S(0.3 , 500000)  ,label = "a = 0.2")
#plt.ylim(-10**(-5) , 10**(-5))
plt.legend()
```

From the picture we can guess $a \in (0.6,0.7]$
