This code is to compute $ f(n) \equiv \frac{\sqrt{n} }{2^{2n}} \binom{2n}{n}$ , we show that it is in fact uniformly bounded .

```python
import numpy as np
import math
```

```python
np.math.factorial(3)
```

```python
def bin(n,k):
    a = np.math.factorial(n)
    b = np.math.factorial(k)
    c= np.math.factorial(n-k)
    if n < k :
        print("error")
    else: 
        return a/(b*c)
```

```python
bin(10,3)
```
